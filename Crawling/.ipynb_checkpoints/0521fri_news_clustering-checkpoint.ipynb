```python
# k - 평균 군집 분석
import pandas as pd
from konlpy.tag import Hannanum
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
```

```python
hannanum = Hannanum()
news_data = pd.read_csv("C:/Users/330-15/Desktop/20_1_computer/SocialWebMining/Crawling/naver_news.csv"
                        , engine = 'python', encoding='utf-8')
news_data.head()
```

```python
# 한나눔 형태소 분석기로 명사만 추출
news_docs = []
for i in news_data['내용']:
    news_docs.append(hannanum.nouns(i))

print (news_docs)
```

```python
# 명사들을 띄어쓰기로 붙여서 열거
for i in range(len(news_docs)):
    news_docs[i] = ' '.join(news_docs[i])

print (news_docs)
```

```python
vec = CountVectorizer()
X = vec.fit_transform(news_docs)

df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
df
```

```python
kmeans = KMeans(n_clusters=3).fit(df)
kmeans.labels_
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(data=principalComponents,
                          columns=['principal component 1', 'principal component 2'])

principalDf.index = news_data['검색어']
principalDf
```

```python
plt.scatter(principalDf.iloc[kmeans.labels_==0, 0], 
            principalDf.iloc[kmeans.labels_==0, 1], 
            s=10, c='red', label='cluster1')
plt.scatter(principalDf.iloc[kmeans.labels_==1, 0], 
            principalDf.iloc[kmeans.labels_==1, 1], 
            s=10, c='blue', label='cluster2')
plt.scatter(principalDf.iloc[kmeans.labels_==2, 0], 
            principalDf.iloc[kmeans.labels_==2, 1], 
            s=10, c='green', label='cluster3')
plt.legend()
```
