```python
# 단어 단위로 토큰화
words = lines.lower()
tokenizer = RegexpTokenizer('[\w]+')
tokens = tokenizer.tokenize(words)

# 불용어 및 한 글자 제거
stop_words = stopwords.words('english')
stopped_tokens = [i for i in list((tokens)) if not i in stop_words]
stopped_tokens2 = [i for i in stopped_tokens if len(i) > 1]
```

```python
# 단어 빈도 카운트
# Series : 1차원
# Matrix : 하나의 자료형만 넣을 수 있는 2차원~n차원
# Data Frame : 여러 자료형을 넣을 수 있는 2차원~n차원
print (pd.Series(stopped_token2).value_counts().head(10))
print ()
print (pd.Series(stopped_token2).value_counts().tail(10))
```

```python
import pandas as pd
from konlpy.tag import Hannanum

hannanum = Hannanum()

f = open('C:/Users/330-15/Desktop/20_1_computer/SocialWebMining/txt/문재인대통령취임연설문.txt', 
         mode='r',
         encoding='utf-8')

lines = f.readlines()
f.close()

temp = []
for i in range(len(lines)):
    temp.append(hannanum.nouns(lines[i]))

def flatten(l):
    flatList = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.apped(elem)
    return flatList

word_list = flatten(temp)
word_list = pd.Series([x for x in word_list if len(x) > 1])
print (word_list.value_counts().head(10))
print ()
print (word_list.value_counts().tail(10))
```

```python
# temp가 리스트안에 또 리스트가 구성되기에 flatten 함수를 정의해서 사용을 해준다.
print (temp)
```

```python
from wordcloud import WordCloud
from collections import Counter

font_path = "C:/Windows/Fonts/malgun.ttf"

wordcloud = WordCloud(
    font_path = font_path,
    width = 800,
    height = 800,
    background_color="white"
)
```

```python
# Counter를 count에게 주고, frequencies를 wordcloud에게 다시 넘겨준다.
count = Counter(stopped_token2)
wordcloud = wordcloud.generate_from_frequencies(count)
```

```python
# 객체에서 다시 array로 바꿔주는 것이다.
array = wordcloud.to_array()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
plt.imshow(array, interpolation="bilinear")
plt.show()
fig.savefig('wordcloud.png')
```
